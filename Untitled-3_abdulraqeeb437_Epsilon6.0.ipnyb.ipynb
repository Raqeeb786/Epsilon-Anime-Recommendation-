```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
revenue_streaming = pd.read_csv('epsilon_work_2.csv')
```

```python
revenue_streaming.head()
```

```python
revenue_streaming.info()
```

```python
# Check for duplicated values
revenue_streaming.duplicated().sum()
```

```python
# Converting data tupes of columns
col= ['Production Budget','Streaming Cost','Merchandise revenue','Streaming revenues','Total revenues']     
for i in col:
    revenue_streaming[i] = revenue_streaming[i].astype(str).str.replace('$', '').str.replace(',', '').str.replace('#REF!','0').astype(float)
revenue_streaming.head()
```

```python
revenue_streaming.describe()
```

```python
for i in revenue_streaming.columns:
    print(i, revenue_streaming[i].nunique())
```

```python
# finding spread of data
col= ['Production Budget','Streaming Cost','Merchandise revenue','Streaming revenues','Total revenues']
for i in col:
    sns.histplot(revenue_streaming[i])
    plt.show()
```

```python
# Filtering the revenue
revenue={'High':0,'Medium':0,'Low':0}
for i in revenue_streaming['Total revenues']:
    if i>=1500000000:
        revenue['High']+=1
    elif 1500000000>i>500000000:
        revenue['Medium']+=1
    else:
        revenue['Low']+=1
revenue=pd.DataFrame(revenue.items(),columns=['Revenue','Count'])
revenue.plot.pie(y='Count',labels=revenue['Revenue'],autopct='%1.1f%%')
plt.ylabel('')  
plt.title('Revenue of Anime')
plt.show()
```

```python
# comaaring the production budget and total revenue
sns.scatterplot(data=revenue_streaming, x='Production Budget', y='Total revenues')
plt.show()
```

```python
# comparing the straeming revneue and merchandise revenue
#sns.scatterplot(data=revenue_streaming, x='Streaming revenues', y='Merchandise revenue')
#plt.show()
sns.histplot(revenue_streaming['Streaming revenues'], color='blue', bins=20, alpha=0.5, label='Streaming revenues')
sns.histplot(revenue_streaming['Merchandise revenue'], color='red', bins=20, alpha=0.5, label='Merchandise revenue')
plt.legend()
plt.show()
```

```python
# comparing the streaming cost and total revenue
sns.scatterplot(data=revenue_streaming, x='Streaming Cost', y='Total revenues', hue='Streaming revenues', palette='viridis')
plt.show()
```

```python
# comparing the streaming cost and streaming revenue
sns.scatterplot(data=revenue_streaming, x='Streaming Cost', y='Streaming revenues', hue='Total revenues', palette='viridis')
plt.show()
```
